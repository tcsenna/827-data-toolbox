### Convers√£o RGB para Grayscale

$$
  Y = 0.299\cdot R + 0.587\cdot G + 0.114\cdot B
$$

````python
# Colorida: 3 canais (RGB)
# Escala de cinza: 1 canal (Cinza)
````

````python
img.shape
````

````python
# Obtendo/separando cada um dos canais da imagem
red = img[:,:,0]
green = img[:,:,1]
blue = img[:,:,2]
````

````python
img_gray = (red * 0.299) + (green * 0.587) + (blue * 0.114)
````

````python
img_gray.shape
````

````python
img_gray.dtype
````

````python
img_gray = img_gray.astype('uint8')
````

````python
plt.imshow(img_gray, cmap='gray')
````

#### M√©todo: m√©dia simples

````python
img_gray_mean = (red * 0.333) + (green * 0.333) + (blue * 0.333)
````

````python
img_gray_mean = img_gray_mean.astype('uint8')
````

````python
plt.imshow(img_gray_mean, cmap='gray')
````

````python
img_diff = np.abs(img_gray.astype('int32') - img_gray_mean.astype('int32')).astype('uint8')
````

````python
img_diff.max()
````

````python
img_diff.min()
````

````python
plt.imshow(img_diff, cmap='gray')
````
